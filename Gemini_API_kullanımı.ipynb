````python
def odtu_bot_chat(user_query):
    global SYSTEM_PROMPTU, ODTU_INFO, client, types
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        config=types.GenerateContentConfig(
            system_instruction=SYSTEM_PROMPTU + ODTU_INFO,
            temperature=0.1
        ),
        contents=user_query,
    )
    return response.text
````
