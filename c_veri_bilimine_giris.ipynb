**Seaborn Kütüphanesi**

```python
import seaborn as sns
```

```python
# gun ve saate gore ısı haritası olusturalım
df_a = df[df['Branch'] == 'A']
df_a
```

```python
hm_data = pd.crosstab(df_a['Day'],df_a['Time'])
hm_data
```

```python
# gunleri siralayalim
hm_data = hm_data.reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
hm_data
```

```python
sns.heatmap(hm_data, annot=True , cmap='YlGnBu') # colormap => YlGnBu: yellow-green-blue
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid') # grafik stili
plt.figure(figsize=(15,8)) # grafik ebatları

sns.heatmap(hm_data, annot=True , cmap='YlGnBu')
plt.title('Gün - Saat İşlem Yoğunluğu')
plt.xlabel('Saat' , fontsize=12)
plt.ylabel('Gün' , fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```
