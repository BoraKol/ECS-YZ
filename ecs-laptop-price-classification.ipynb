```python
def convert_storage_to_gb(value):
    # Eğer değer NaN (boş) ise doğrudan döndür
    if pd.isna(value):
        return np.nan
    
    # Küçük harfe çevir ve boşlukları temizle
    value = str(value).lower().strip()
    
    # Sayısal kısmı ayıkla (regex ile)
    import re
    number_match = re.search(r'(\d+)', value)
    if not number_match:
        return np.nan
    
    number = int(number_match.group(1))
    
    # Birim kontrolü ve dönüşüm
    if 'tb' in value:
        return number * 1024
    else:
        # GB veya birim belirtilmemişse direkt sayıyı döndür
        return number
```

```python
def calculate_pixels(res):
    if pd.isna(res):
        return np.nan
    try:
        # 'x' işaretine göre ayır ve iki sayıyı çarp
        parts = str(res).lower().split('x')
        width = int(parts[0].strip())
        height = int(parts[1].strip())
        
        # Hatalı girişleri (1921, 1922 vb.) 1920'ye yuvarla
        if 1910 <= width <= 1940: width = 1920
        
        return width * height
    except:
        return np.nan

df['Toplam_Piksel'] = df['Max Ekran Çözünürlüğü'].apply(calculate_pixels)
```

```python
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Sıralamayı biz belirliyoruz (Küçükten büyüğe)
fiyat_siralamasi = [['Çok Düşük', 'Düşük', 'Orta', 'Yüksek', 'Çok Yüksek']]

encoder = OrdinalEncoder(categories=fiyat_siralamasi)

# Fiyat sütununu dönüştür (0, 1, 2, 3, 4 değerlerini alacaktır)
df['Fiyat'] = encoder.fit_transform(df[['Fiyat']])
```

```python
X = df.drop("Fiyat", axis = 1)
y = df["Fiyat"]
```

```python
from sklearn.model_selection import train_test_split 
X_train , X_test , y_train , y_test = train_test_split(X,y, test_size = 0.25, random_state = 15)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Az seçenekli olanlar (OHE için ideal)
ohe_cols = ["Ekran Panel Tipi", "İşletim Sistemi"]
# Çok seçenekli olanlar (Ordinal için daha güvenli)
ord_cols = ["İşlemci", "Ürün Modeli"]

preprocessor = ColumnTransformer(transformers=[
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ohe_cols),
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), ord_cols)
], remainder="passthrough")

# Dönüştürme işlemleri
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_transformed)
X_test_scaled = scaler.transform(X_test_transformed)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train_scaled,y_train)
```

```python
y_pred = rfc.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report

print("Accuracy: ", accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
