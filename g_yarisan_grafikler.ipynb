```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_secilen = df[['date','Turkey' , 'Italy' , 'Germany' , 'United Kingdom' ,  'China' , 'US']]
df_secilen
```

```python
df_secilen.set_index("date", inplace = True)
```

```python
df_secilen.plot()
```

```python
sum = df_secilen.sum(axis = 0)
sum
```

```python
sum_list = df_secilen.sum(axis=1)
sum_list # 6 ulkedeki toplam vaka sayısı
```

```python
# gunluk alti ulkedeki toplam vaka sayisi
sum_daily = df_secilen.cumsum(axis=0)
sum_daily
```

```python
sum_daily.plot()
```
