## Gradio Form Arayüzünü Oluştur

### Subtask:
Kullanıcı bilgilerini (örn. Ad, E-posta, Yaş) toplamak için metin kutuları (gr.Textbox) ve diğer gerekli bileşenlerle bir Gradio arayüzü tasarlayın.


**Reasoning**:
To design the Gradio interface as requested by the subtask, I will import the `gradio` library, define a placeholder function for processing form submissions, and then create the Gradio interface with `gr.Textbox` components for 'Ad', 'E-posta', and 'Yaş', setting appropriate types and labels.



**Reasoning**:
The previous code failed because `type="number"` is not a valid type for `gr.Textbox` in Gradio. The error message explicitly states that the `type` must be one of "text", "password", or "email". To fix this, I will remove the `type="number"` attribute from the `age_input` Textbox.



**Reasoning**:
The subtask of defining the Gradio interface components is complete. The next logical step is to implement the function that saves the submitted user data to the CSV file. This involves modifying the `submit_form` function to append the name, email, and age to the specified CSV file.



```python
import gradio as gr
import csv
import os

csv_file_path = '/content/drive/MyDrive/YZ Eğitimleri/2026/VBECS11-25/Form.csv'

def submit_form(full_name, email, age, gender, department, employment_status):
    try:
        with open(csv_file_path, 'a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow([full_name, email, age, gender, department, employment_status])
        return "Form başarıyla gönderildi ve bilgiler kaydedildi!"
    except Exception as e:
        return f"Bilgiler kaydedilirken bir hata oluştu: {e}"

with gr.Blocks() as demo:
    gr.Markdown("# Kullanıcı Bilgileri Formu")
    with gr.Column():
        full_name_input = gr.Textbox(label="Ad Soyad", placeholder="Adınızı ve Soyadınızı girin")
        email_input = gr.Textbox(label="E-posta", placeholder="E-posta adresinizi girin", type="email")
        age_input = gr.Textbox(label="Yaş", placeholder="Yaşınızı girin")
        gender_input = gr.Dropdown(label="Cinsiyet", choices=["Erkek", "Kadın", "Belirtmek İstemiyorum"], value="Belirtmek İstemiyorum")
        department_input = gr.Textbox(label="Bölüm", placeholder="Bölümünüzü girin")
        employment_status_input = gr.Textbox(label="İş Durumu", placeholder="İş durumunuzu girin (Öğrenci, Çalışan, İşsiz vb.)")
        submit_button = gr.Button("Gönder")
    output_message = gr.Textbox(label="Durum")

    submit_button.click(
        submit_form,
        inputs=[full_name_input, email_input, age_input, gender_input, department_input, employment_status_input],
        outputs=output_message
    )

# The Gradio interface will be launched in the next step after implementing the save logic.
print("Gradio interface with extended CSV save logic defined.")
```

**Reasoning**:
The Gradio interface has been defined and the CSV saving logic is integrated. To complete the subtask of building a Gradio web application, the final step is to launch the Gradio interface so it can be accessed and used.



```python
demo.launch()
```
